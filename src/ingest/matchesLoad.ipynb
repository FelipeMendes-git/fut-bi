```python
GAME_ID = 18235
```

```python
import json

match_path = f"../../data/raw/matches/{GAME_ID}.json"
with open(match_path, encoding="utf-8") as f:
    matches = json.load(f)

m = matches[0]
home = m["home_team"]
away = m["away_team"]
home_manager = (home.get("managers") or [{}])[0].get("name")
away_manager = (away.get("managers") or [{}])[0].get("name")
stadium = m.get("stadium", {}).get("name")
```

```python
from conexao import conectar

sql = """
    INSERT INTO games (
        game_id,
        team_1_id, team_1_name,
        team_2_id, team_2_name,
        home_score, away_score,
        home_manager, away_manager,
        stadium
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

row = (
    m["match_id"],
    home["home_team_id"],
    home["home_team_name"],
    away["away_team_id"],
    away["away_team_name"],
    m["home_score"],
    m["away_score"],
    home_manager,
    away_manager,
    m.get("stadium", {}).get("name"),
)

conn = conectar()
try:
    cursor = conn.cursor()
    cursor.execute(sql, row)
    conn.commit()
    print("Jogo inserido.")
finally:
    cursor.close()
    conn.close()
```
